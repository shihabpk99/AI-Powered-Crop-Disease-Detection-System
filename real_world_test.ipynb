```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, BatchNormalization, Dropout
from tensorflow.keras.models import Model

print("🧹 1. Waking up the Heavyweight Champion...")
tf.keras.backend.clear_session()

# ==========================================
# 1. REBUILD & LOAD DENSENET121
# ==========================================
weights_dir = r"D:\00_Thesis_Final_Models"
model_weights_path = os.path.join(weights_dir, "DenseNet121_Fresh_Phase2_FineTuned.h5")

base_dense = DenseNet121(weights=None, include_top=False, input_shape=(224, 224, 3))
x_dense = GlobalAveragePooling2D()(base_dense.output)
x_dense = BatchNormalization()(x_dense)
x_dense = Dense(256, activation='relu')(x_dense)
x_dense = Dropout(0.4)(x_dense)
preds_dense = Dense(15, activation='softmax')(x_dense)

model = Model(inputs=base_dense.input, outputs=preds_dense)
model.load_weights(model_weights_path)
print("✅ DenseNet121 Loaded and ready for real-world data!")

# ==========================================
# 2. THE 15 DISEASE CLASSES (Alphabetical Order)
# ==========================================
class_names = [
    'potato_bacteria', 'potato_early_blight', 'potato_fungi', 'potato_healthy', 'potato_late_blight',
    'rice_blast', 'rice_brown_spot', 'rice_healthy', 'rice_sheath_blight', 'rice_tungro_virus',
    'wheat_black_point', 'wheat_blast', 'wheat_fusarium_foot_rot', 'wheat_healthy', 'wheat_leaf_blight'
]

# ==========================================
# 3. TEST THE REAL WORLD PHOTOS
# ==========================================
real_world_dir = r"D:\Real_World_Rice"
print(f"\n🔬 2. Scanning folder: {real_world_dir}...\n")

# Loop through all files in your new folder
for filename in os.listdir(real_world_dir):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        img_path = os.path.join(real_world_dir, filename)
        
        # Load and preprocess the image exactly how the model expects it
        img = image.load_img(img_path, target_size=(224, 224))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array /= 255.0  # Rescale
        
        # Make the prediction
        predictions = model.predict(img_array, verbose=0)
        predicted_class_index = np.argmax(predictions[0])
        predicted_disease = class_names[predicted_class_index]
        confidence = predictions[0][predicted_class_index] * 100
        
        # Print the result
        print(f"📸 Image: {filename}")
        print(f"   🦠 Diagnosis: {predicted_disease.upper()} ({confidence:.2f}% confidence)\n")

print("✅ Real-world field test complete!")
```
